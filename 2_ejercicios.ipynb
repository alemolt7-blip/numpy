## Ejercicio 1: Explorar propiedades de la imagen

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('beagle.jpg')

# Shape de la imagen
print('Shape:', img.shape)

# Tipo de dato
print('Dtype:', img.dtype)

# Valor máximo y mínimo de píxel
print('Valor máximo:', img.max())
print('Valor mínimo:', img.min())

# Valor del píxel en la posición [10, 10]
print('Pixel [10,10]:', img[10, 10])
```

## Ejercicio 2: Mostrar la imagen original

```python
plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.title('Imagen Original - Beagle')
plt.axis('off')
plt.show()
```

## Ejercicio 3: Convertir la imagen a escala de grises

```python
# Convertir a escala de grises usando la fórmula estándar
img_gris = np.dot(img[..., :3], [0.2989, 0.5870, 0.1140])

print('Shape imagen gris:', img_gris.shape)

plt.figure(figsize=(6, 6))
plt.imshow(img_gris, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.axis('off')
plt.show()
```

## Ejercicio 4: Separar canales RGB

```python
# Separar los canales de color
canal_r = img[:, :, 0]
canal_g = img[:, :, 1]
canal_b = img[:, :, 2]

print('Promedio canal Rojo:', canal_r.mean().round(2))
print('Promedio canal Verde:', canal_g.mean().round(2))
print('Promedio canal Azul:', canal_b.mean().round(2))

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(canal_r, cmap='Reds')
axes[0].set_title('Canal Rojo')
axes[0].axis('off')
axes[1].imshow(canal_g, cmap='Greens')
axes[1].set_title('Canal Verde')
axes[1].axis('off')
axes[2].imshow(canal_b, cmap='Blues')
axes[2].set_title('Canal Azul')
axes[2].axis('off')
plt.tight_layout()
plt.show()
```

## Ejercicio 5: Recortar una sección de la imagen

```python
# Recortar la parte central de la imagen
recorte = img[256:768, 256:768]

print('Shape del recorte:', recorte.shape)

plt.figure(figsize=(5, 5))
plt.imshow(recorte)
plt.title('Recorte Central (512x512)')
plt.axis('off')
plt.show()
```
